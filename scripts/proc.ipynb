```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cols = ["Time", "Shard", "URL"]
df = pd.DataFrame()
f = open('/tmp/ray/session_latest/restore.log', 'r')
limit = 1000

for i, line in enumerate(f.readlines()):  
    if i >= limit:
        break   
    restore_info = json.loads(line)
    time = pd.to_datetime(restore_info['time'])
    shard = restore_info['url'].split('/')[3]
    url = restore_info['url_with_offset']
    data = {
        "Time": time,
        "Shard": shard,
        "URL": url
    }
    df = df.append(data, ignore_index = True)
```

```python
df.groupby('Shard').size().plot.bar()
```
